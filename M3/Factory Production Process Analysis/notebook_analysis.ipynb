```python
import pandas as pd
import streamlit as st
df = pd.read_csv("production_quality_dataset.csv")
print(df.head(5))
print(df.info())
df["scrap_rate"] = df["defective_units"] / df["units_produced"].replace(0, pd.NA)
df["scrap_rate_percent"] = (df["scrap_rate"] * 100).round(2)
print(df["scrap_rate_percent"])
highest_scrap_rate_batch = df.sort_values("scrap_rate", ascending = False).iloc[0]
print(highest_scrap_rate_batch)
df1 = df.groupby("machine")["scrap_rate"].mean()
print(df1)
df2 = df.groupby("shift")["scrap_rate"].mean()
print(df2)
```
